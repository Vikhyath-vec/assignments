```python
import numpy as np
import pandas as pd
import random
random.seed(1)
```

```python
data = pd.read_excel("Residential-Building-Data-Set.xlsx", header=1)
```

```python
X = data.iloc[:, :107].to_numpy()
Y = data.iloc[:, 107:].to_numpy()

id_list = list(range(len(X)))
random.shuffle(id_list)
id_train = id_list[0:186]
print(len(id_train))
id_test = id_list[186:]
print(len(id_test))
```

```python
X_train = X[id_train]
print(X_train.shape)
Y_train = Y[id_train]
print(Y_train.shape)

X_test = X[id_test]
print(X_test.shape)
Y_test = Y[id_test]
print(Y_test.shape)
```

```python
print(id_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import explained_variance_score
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train, Y_train)
```

```python
Y_pred = linear_reg.predict(X_test)
# print(Y_pred)
# print(explained_variance_score(Y_test, Y_pred))
print("-----Linear Regression-----")
print("Explained Variances:")
print(explained_variance_score(Y_test[:, 0], Y_pred[:, 0]), explained_variance_score(Y_test[:, 1], Y_pred[:, 1]))
a = linear_reg.coef_
print(a[0])
print(a[1])
```
