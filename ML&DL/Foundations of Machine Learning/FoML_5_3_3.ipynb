```python
import numpy as np
import scipy.io
import pandas as pd
import random
from sklearn.preprocessing import StandardScaler, normalize
import matplotlib.pyplot as plt
from sklearn import datasets
random.seed(0)
```

```python
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# using the attribute information as the column names
col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
iris =  pd.read_csv(csv_url, names = col_names)
```

```python
X = iris.iloc[:, :4]
# Y = iris.iloc[:, 4]
irisb = datasets.load_iris()
Y = irisb.target
# Y = pd.get_dummies(Y)
print(X)
print(Y)
```

```python
X = X.to_numpy()
# Y = Y.to_numpy()
```

```python
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
```

```python
from sklearn.svm import SVC

C_list = [0.01, 0.1, 1, 10, 100]
for c in C_list:
  svc = SVC(kernel='linear', C=c)
  svc.fit(X[:, :2], Y)
  Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])

  # Put the result into a color plot
  Z = Z.reshape(xx.shape)
  plt.figure()
  plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
  plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Spectral, edgecolors='black')
  plt.xlabel('Sepal length')
  plt.ylabel('Sepal width')
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
  plt.xticks(())
  plt.yticks(())
  plt.title("SVC with linear kernel (C = {})".format(c))
  plt.show()
```

```python
for c in C_list:
  svc = SVC(kernel='poly', degree=2, C=c)
  svc.fit(X[:, :2], Y)
  Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])

  # Put the result into a color plot
  Z = Z.reshape(xx.shape)
  plt.figure()
  plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
  plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Spectral, edgecolors='black')
  plt.xlabel('Sepal length')
  plt.ylabel('Sepal width')
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
  plt.xticks(())
  plt.yticks(())
  plt.title("SVC with polynomial kernel of degree 2 (C = {})".format(c))
  plt.show()
```

```python
for c in C_list:
  svc = SVC(kernel='poly', degree=3, C=c)
  svc.fit(X[:, :2], Y)
  Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])

  # Put the result into a color plot
  Z = Z.reshape(xx.shape)
  plt.figure()
  plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
  plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Spectral, edgecolors='black')
  plt.xlabel('Sepal length')
  plt.ylabel('Sepal width')
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
  plt.xticks(())
  plt.yticks(())
  plt.title("SVC with polynomial kernel of degree 3 (C = {})".format(c))
  plt.show()
```

```python
gamma_list = [0.01, 0.1, 1, 10, 100]
for c in C_list:
  for gamma in gamma_list:
    svc = SVC(kernel='rbf', gamma=gamma, C=c)
    svc.fit(X[:, :2], Y)
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Spectral, edgecolors='black')
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title("SVC with RBF kernel(gamma = {}, C = {})".format(gamma, c))
    plt.show()
```
