```python
import numpy as np
import pandas
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
n_samples = 500
random_state = 0 #this is for reproducibility
#check 90 later on

stds = [1.5, 2.5, 0.75]

X, Y , centers = datasets.make_blobs(
    n_samples=n_samples, cluster_std=stds, random_state=random_state, return_centers = True
)

print(centers)

print(np.square(stds))
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components = 3, covariance_type = 'spherical')
gmm.fit(X)
Y_pred = gmm.predict(X)

print(gmm.means_)

print(gmm.covariances_)
```

```python
def plotpoints(X, Y):
  fig = plt.figure()

  plt.scatter(X[:, 0], X[:, 1], c=Y, cmap = plt.cm.brg)
  plt.show()
```

```python
plotpoints(X, Y)
```

```python
plotpoints(X, Y_pred)
```
