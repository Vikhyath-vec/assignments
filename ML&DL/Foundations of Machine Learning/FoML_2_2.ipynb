# Converting to numpy

```python
X = df1.to_numpy()
print(X.shape)

for j in range(X.shape[0]):  # verifying that no data was lost
    for ix, i in enumerate(df1.iloc[j]):
        if i!=X[j, ix]:
            print("Issue")
print("Done")
```

```python
Y = df2.to_numpy()  # then you may need to convert output label 2 to 0 or -1 if the library to-be-used demands so.
print(Y.shape)
```

```python
ones = np.ones((126, 1))
X2 = np.square(X)
X = np.concatenate((X, X2), axis=1)
X = np.concatenate((ones, X), axis=1)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X, Y)
```

```python
id_list = list(range(len(X)))
random.shuffle(id_list)
id_train = id_list[0:63]
print(len(id_train))
id_test = id_list[63:]
print(len(id_test))
```

```python
X_train = X[id_train]
print(X_train.shape)
Y_train = Y[id_train]
print(Y_train.shape)
```

```python
X_test = X[id_test]
print(X_test.shape)
Y_test = Y[id_test]
print(Y_test.shape)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
per = Perceptron()
per.fit(X_train, Y_train.ravel())
Y_pred = per.predict(X_test)
print(accuracy_score(Y_test, Y_pred))
```

```python
log_reg = LogisticRegression(solver='lbfgs', max_iter=1000)
log_reg.fit(X_train, Y_train.ravel())
Y_pred = log_reg.predict(X_test)
print(accuracy_score(Y_test, Y_pred))
```
